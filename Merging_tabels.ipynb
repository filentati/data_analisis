## Объединение таблиц

```python
import pandas as pd
```

```python
raw_data ={
    'subject_id': ['1', '2', '3', '4', '5'],
    'first_name': ['Alex', 'Any', 'Anna', 'Archi', 'Amanda'],
    'last_name': ['Anderson', 'Askermen', 'Ali', 'Arlov', 'Asonlo']}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_a.index = [1, 2, 3, 4, 5]
df_a
```

```python
raw_data ={
    'subject_id': ['6', '7', '8', '4', '5'],
    'first_name': ['Bob', 'Bill', 'Bella', 'Borya', 'Blum'],
    'last_name': ['Bond', 'Bonce', 'Break', 'Blomov', 'Black']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b.index = [2, 3, 4, 5, 6]
df_b
```

```python
raw_data = {
    'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'], 
    'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_n = pd.DataFrame(raw_data, columns = ['subject_id', 'test_id'])
df_n
```

```python
# метод concat импользуется для объединения таблиц по вертикальной или горизонтальной оси
df_new = pd.concat([df_a, df_b], axis = 0) # axis = 0 объединяет таблицы построчно 
df_new
```

```python
df_new = pd.concat([df_a, df_b], axis = 1) # axis = 1 объединяет таблицы по колонкам, по индесам
df_new
```

```python
df_new = pd.concat([df_a, df_b], axis = 1, join = 'inner') # с ключевым словом inner остаются строкиБ в которых присутсвуют индесы как в первой так и втрой таблицы 
df_new
```

```python
df_new = pd.concat([df_a, df_b], axis = 0, join = 'outer') 
df_new
```

Метод Join основан на объединение таблиц через индесы с помощью параметра how = ['left', 'right', 'inner', 'outer']

```python
df_a.join(df_b, rsuffix ='_', how = 'left') # с  помощь left у нас берутся все строки из датафрейма 1 и к ним добавляются из датафрейма2 
```

Merge используется для объединения таблиц по любым колонкам 

```python
df_new
```

```python
pd.merge(df_new, df_n, on = 'subject_id') # при этом subject_id не дублируется в новой таблице
```

```python
df_nn = df_n.rename(columns = { 'subject_id': 'id_subject'})
pd.merge(df_new, df_nn, left_on = 'subject_id', right_on = 'id_subject')
```

```python
# если мы хотим объединить по индексам 
pd.merge(df_a, df_b, right_index = True, left_index = True)
```
